```python
import pandas as pd
import numpy as np
```

```python
sal=pd.read_csv(r"Salaries.csv",low_memory=False)
```

```python
sal=sal.drop_duplicates()  ## as there are chances of duplicate data
```

```python
sal.drop(sal[sal["BasePay"]== 'Not Provided'].index,inplace=True)
```

```python
sal
```

```python
sal.describe()
```

```python
sal.info()
```

```python
sal=sal.drop(columns='Notes')
```

```python
sal['Benefits'] = pd.to_numeric(sal['Benefits'], errors='coerce').astype('float64')
sal['BasePay'] = pd.to_numeric(sal['BasePay'], errors='coerce').astype('float64')
sal['OvertimePay'] = pd.to_numeric(sal['OvertimePay'], errors='coerce').astype('float64')
sal['OtherPay'] = pd.to_numeric(sal['OtherPay'], errors='coerce').astype('float64')
sal['TotalPay'] = pd.to_numeric(sal['TotalPay'], errors='coerce').astype('float64')
```

How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?)

How many people have the word Chief in their job title?
